```python
import pyspark
```

```python
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("Log Processing") \
    .getOrCreate()
```

```python
df = spark.read \
    .option("header", "true") \
    .option("mode", "DROPMALFORMED") \
    .option("delimiter","\t") \
    .csv("sample.log")
```

```python
df.show(5)
```

```python
df = spark.read.load("sample.log",
                     format="csv", 
                     sep="\t", 
                     inferSchema="true", 
                     header="true")
```

```python
df.show(5)
```

```python
df = spark.sql("SELECT * FROM text.`sample.log`")
```

```python
df.show(5)
```
